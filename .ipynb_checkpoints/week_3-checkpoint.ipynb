Display a table and a chart for the top 5 download apps with smallest size and highest downloading number
* Display a table and a chart for top 5 download apps in each category free and paid (if there is paid)
* Display a sunburst plot grouping by categories

```python
import numpy as np
from matplotlib import pyplot as plt
import os
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('googleplaystore.csv')
df
```

```python
sizedrop = df[df.Size == 'Varies with device'].index
sizedrop
```

```python
df_size = df.drop(sizedrop)
df_size
```

```python
dfsize_k = df_size[df_size['Size'].str.contains('k')]
dfsize_k['Size']=dfsize_k.Size.replace('k', '', regex=True).astype(float)
dfsize_k['Size']
```

```python
smallest_sizes = dfsize_k.sort_values('Size').head().sort_values('Installs', ascending=False)
smallest_sizes
```

```python
barchart_size_installs =px.bar(
    smallest_sizes, x='Size', y='Installs',
    title='Top 5 download apps with smallest size and highest Installs'.title(),
    text='Installs', hover_name='Size', color='App',
    labels={'Installs': 'highest downloading number'.title()
)
```

### Display a table and a chart for top 5 download apps in each category free and paid (if there is paid)

```python
df = df[(df['Price'] != 'Everyone')]
```

```python
df['Price'] = df.Price.replace('[$+,]', '', regex=True).astype(float)
df['Price']
```

```python
df["Installs"] = df["Installs"].replace('[\+,]', '', regex = True).astype(float)
```

```python
free_type = df[df.Type == 'Free'].groupby(['Category', 'App'], as_index=False)[['Installs']].sum().sort_values('Installs').tail()
free_type
```

```python
topfreetype = px.bar(free_type, x='App', y='Installs', color='Category',
                    template='plotly_dark', width=800, height=600, title='top 5 download free apps'.title(), 
                    labels={'Installs': 'No. of download', 'Category': 'Categories'}
)
```

```python
#Paid Type 2. Display a table and a chart for top 5 download apps in each category free and paid (if there is paid)
paid_type = df[df.Type == 'Paid'].groupby(['Category', 'App'], as_index=False)[['Installs']].sum().sort_values('Installs').tail()
paid_type
```

### Display a sunburst plot grouping by categories

```python
px.sunburst(df, path=['Category', 'App', 'Reviews'], values='Installs',
               color='Size', hover_data=['Reviews'],
               hover_name='Category', title='Google Play Store Categories'
           )
```
