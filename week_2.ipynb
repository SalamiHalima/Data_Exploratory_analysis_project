###    Plot a bar chart for categories with the total number of installing numbers in each category

```python
df['Installs'].unique()
```

```python
df['Installs'] = df.Installs.replace('[/+,]', '', regex= True).astype(float)
```

```python
df.groupby('Category')['Installs'].count().sort_values().plot(kind='bar', ylabel='Installs', title=' the total number of installing numbers'.title(), color = 'green', figsize=(10,5))
```

###    Plot a bar chart for the total prices of each paid app in each category ( the sum of all prices in the same category)

```python
df.Price.unique()
```

```python
df['Price'] = df.Price.replace('[/+,$]', '', regex=True).astype(float)
```

```python
df.groupby('Category')['Price'].sum().sort_values().plot(kind='bar', ylabel='Price', title='the total number of installing numbers'.title(), color = 'green', figsize=(10,5))
```

### Plot a bar chart of the total profit of each category by multiplying the price by the number of installs

```python
df['total_profit'] = df.Price * df.Installs
df.groupby('Category')['total_profit'].sum().sort_values().plot(kind='bar', ylabel='total_profit', title='the total profit of each category'.title(), color = 'green', figsize=(10,5))
```
